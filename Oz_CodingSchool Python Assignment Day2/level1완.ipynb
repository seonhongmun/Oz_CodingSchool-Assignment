__[문제2]__


변수 num1에 15, num2에 4를 할당하고, 두 변수를 더한 값, 뺀 값, 곱한 값, 나눈 값을 각각 출력하세요.



<br>

```python
num1, num2 = 15, 4
print(num1 + num2) #더하기
print(num1 - num2) #빼기
print(num1 * num2) #곱하기
print(num1 / num2) # 나누기 / 소수점까지
print(num1 // num2) # 나누기 / 몫만
print(num1 % num2) # 나누기 / 나머지만
```

__[문제5]__

str = "야호야호야호호야호야야오햐오햐오햐오"

str을 3가지 이상의 다른 문자열 슬라이싱을 이용하여 확인해보세요!

출력 예)

처음 5글자: 야호야호야


```python
str = "야호야호야호호야호야야오햐오햐오햐오"
print(len(str)) #18
print(str[0:5])
print(str[10:2:-1])
print(str[5:18])
print(str[::-1])
print(str[::3])
```
