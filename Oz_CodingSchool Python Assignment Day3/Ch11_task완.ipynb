<span style="font-size:80%">[문제3]<br><br></span>
아래 코드를 실행시켰을 때 발생하는 상황과 원인을 간략히 설명해주세요

```
바구니 = ( "상추", "깻잎", "삼겹살")
바구니[1] = "고사리"
```

````python
바구니 = ( "상추", "깻잎", "삼겹살")
print(바구니)
바구니 = list(바구니) #list 형변환
print(type(바구니))  #타입 확인
바구니[1] = "고사리" # list화 한 깻잎을 고사리로 변경
# print(list화)  # 변경이 되었는지 확인
바구니 = tuple(바구니)
print(바구니)
# 바구니[1] = "고사리"
````

<span style="color:red">답변</span>

오류가 발생한다.  
원인으로는 바구니는 print(type(바구니))를 해보게 되면 tuple의 타입으로 나오게 되는데   
tuple은 불변이기 때문에 바구니[1] = "고사리"는 사용할수가 없다.
tuple은 추가, 수정, 삭제가 불가능한 불변의 변수이기 때문이다.
바구니[1] = "고사리"의 명령어를 실행시키고 싶다면 list(바구니)를 통해 형변환을 해주고 고사리를 넣고 다시 tuple로 형변환 하는 방법이 있다.
