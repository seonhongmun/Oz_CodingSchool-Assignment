__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
print(nums[-1])
print(nums[0])
print(nums[4])

## --------------------------

max_1 = max(nums)
min_1 = min(nums)
average = sum(nums)//len(nums)

print("최대값", max_1)
print("최소값", min_1)
print("평균", average)
````
