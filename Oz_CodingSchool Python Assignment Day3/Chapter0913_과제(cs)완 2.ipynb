[문제2]

Python의 set에 대해 설명해주세요

set이란 중복되는 원소가 없이, 순서에 상관없는 데이터들의 묶음이다.   
데이터의 중복을 허용하지 않기 때문에 중복되는 원소가 있다면 하나만을 저장한다.  
또한, 순서가 없기 때문에 리스트처럼 인덱스 번호를 사용하여 특정 값을 접근할 수 없다.

[문제3]

Python의 list는 연결리스트(linked list)인가요?

파이썬의 리스트는 연결리스트가 아니다.
list는 동적배열을 사용하여 구현되어있으며 동적배열은 일정량의 메모리를 할당받아 가져오고

만약 4개의 메모리를 다 사용했다면 8, 11, 15의 메모리를 추가 할당받아 추가, 삭제,
수정 등 여러가지 연산을 수행한다.  

연결리스트란 선형 자료구조 중 배열과 함께 가장 기본이 되는 개념으로,

"물리 메모리 어딘가에 흩뿌려진 노드(Node)들이 서로 연결된 형태로 구성된 자료구조"로 의미하기에 list와 연결리스트와는 다른 개념이다.

[문제4]

딕셔너리 자료형에 대해 설명해주세요.

딕셔너리란 단어 그대로 "사전"이란 뜻이고 Key가 되는 단어와 Value가 되는 의미하는 단어를   
한쌍으로 이루어진 가변적이고 순서가 없는 자료구조이다.  
효율적인 데이터 검색을 위해 해시 테이블 구조를 사용하며, 중복되지 않는 고유한 키를 이용해   
데이터에 접근하며, 중괄호({})를 사용하여 딕셔너리를 생성하거나 dict() 함수로 생성이 가능하다.   
또한, 딕셔너리는 데이터 조작에 유용한 여러 메서드를 제공한다.
