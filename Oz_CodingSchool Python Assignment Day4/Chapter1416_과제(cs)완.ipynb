[문제1]

파이썬 2.7 버전과 파이썬 3에서 range 차이를 설명해주세요

파이썬 2.7 range함수는 실제 리스트를 생성하여 큰 범위로 갈 수록 메모리 사용량이 높기에 주의해야한다.   
파이썬 3 range함수는 이터레이터(iterator)를 반환한다.
이터레이터란 파이썬에서 반복을 지원하기 위한 객체이며 내부적으로 '_ _ iter _ ()' 와 ' _ next _ _()'의 메서드를 가지고 있다. 주로 for루프와 같이 반복문에서 사용하며, 요소를 필요할 때마다 생성하므로, 메모리를 효율적으로 사용할 수 있는 기능을 말한다.
그렇기에 range함수를 호출해도 결과를 바로 리스트에 반환하지 않으므로, 필요한 경우에만 리스트로 변환하여 사용할 수 있다.
