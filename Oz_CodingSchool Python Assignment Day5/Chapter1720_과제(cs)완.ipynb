[문제3]

클래스 상속에 대해 설명해주세요

상속이란 두 클래스 사이에 부모-자식 관계를 설정하는 것.  
부모와 자식이 같은 이름의 메서드를 갖고 있다면 mro에서 제일 빠른 순번의 것이 먼저 호출된다. 즉, 자식에서 오버라이딩 한 메서드가 호출된다. 이런 원리를 통해서 오버라이딩이 가능한 것이다.   
mro : 어떤 클래스의 상속 관계를 보여준다.   
ininstance : 어떤 인스턴스가 주어진 클래스의 인스턴스인지 알려준다.  
inssubclass : 어떤 클래스가 다른 클래스의 자식 클래스인지 알려준다.  

예) 벤츠-자동차 관계에서 벤츠는 자동차지만 자동차는 벤츠가 아니다. > 자동차는 벤츠의 부모다. > 벤츠는 자동차를 상속받는다. > 상속은 반복을 줄여준다.
