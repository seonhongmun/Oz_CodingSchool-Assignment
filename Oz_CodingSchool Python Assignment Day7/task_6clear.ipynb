requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

```python
# 터미널에 pip3 install requests (pip3 -> pip변경되어있을시 pip install requests)
import requests

web = requests.get('https://kream.co.kr')
print("상태 코드: ", web.status_code)
print("HTML 내용 (100글자) :", web.text[:100])
```
