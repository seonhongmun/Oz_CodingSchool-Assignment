[문제]

'Account' 클래스를 작성하세요.

이 클래스는 소유자(owner)와 잔액(balance)을 속성으로 가지며, 잔액은 프라이빗 변수로 선언합니다.

'deposit'와 'withdraw' 두 개의 메소드를 통해 잔액을 조작할 수 있도록 하세요.

이 때, 잔액은 음수가 될 수 없도록 체크해야 합니다. 'Account' 객체를 생성하고,<br>
입금 및 출금 작업을 시뮬레이션한 후, 최종 잔액을 출력하세요.

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner # 계좌 소유자 이름
        self.balance = balance # 계좌 잔액 (프라이빗 변수)

    def deposit(self, amount):
        """입금 메소드: 금액을 입금하고 현재 잔액을 출력"""
        if amount > 0: # 입금 금액이 0보다 큰지 확인
            self.balance += amount # 잔액에 입금 금액 추가
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.balance}원")
        else:
            print("입금 금액은 0보다 커야 합니다.") # 잘못된 금액 처리

    def withdraw(self, amount):
        """출금 메소드: 금액을 출금하고 현재 잔액을 출력"""
        if amount < 0: # 잔액 초과 출금 방지
            print("출금 금액은 0보다 작을 수 없습니다.")
        elif amount > self.balance: #잔액 초과 출금방지
            print(f'잔액이 부족합니다. 현재금액은 {self.balance}원 입니다. ')
        else:
            self.balance -= amount  # 잔액에서 출금 금액 차감
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance}원")

    def get_balance(self):
        return self.balance  # 현재 잔액 반환
# Account 객체 생성
account = Account("홍길동", 50000) #이름, 잔액 설정

# 입금 및 출금
account.deposit(103000) # 103000원 입금
account.withdraw(20000) # 20000원 출금
account.withdraw(1000000) # 1000000원 출금 (과출금 방지 확인)
# 최종 잔액 출력
print(f'계좌 소유자: {account.owner} "잔액: {account.get_balance()}원 입니다. ')
```
