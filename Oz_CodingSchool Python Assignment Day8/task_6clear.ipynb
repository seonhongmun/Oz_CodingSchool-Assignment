[문제]

'Vehicle'이라는 기본 클래스를 선언하세요.

이 클래스는 make (제조사)와 model (모델명) 두 개의 속성을 갖고, 생성자(__init__)를 통해 이 두 속성을 초기화합니다. 'display_info'라는 메소드를 추가하여, "이 차량은 [make]에서 제조한 [model]입니다."라는 메시지를 반환하도록 하세요.

'Truck'과 'Car' 두 개의 하위 클래스를 생성하고, 각 클래스는 'Vehicle' 클래스로부터 상속받습니다.

'Truck' 클래스에는 추가로 'payload' (적재량) 속성을, 'Car' 클래스에는 'seats' (좌석 수) 속성을 추가합니다. 각각의 하위 클래스에서 'display_info' 메소드는 기본 클래스의 메소드를 그대로 사용합니다.

'Truck'과 'Car' 객체 각각을 생성하고, 'display_info' 메소드를 호출하여 정보를 출력하세요.

또한, 각 객체의 추가 속성(적재량, 좌석 수)에 대한 정보도 출력하세요.

이 문제는 클래스의 상속 개념을 이해하고, 기본 클래스의 속성과 메소드를 하위 클래스가 어떻게 활용할 수 있는지 보여줍니다.

오버라이드는 필요하지 않으며, 각 하위 클래스의 고유한 속성을 추가하는 방법에 초점을 맞춥니다.

```python
# 기본 클래스 Vehicle 선언
class Vehicle:
    def __init__(self, make, model):
        """Vehicle 클래스 생성자: 제조사와 모델명을 초기화"""
        self.make = make  # 제조사
        self.model = model  # 모델명

    def display_info(self):
        """Vehicle 클래스 정보 표시 메소드"""
        return f"이 차량은 {self.make}에서 제조한 {self.model}입니다."


# Truck 클래스: Vehicle 클래스를 상속받음
class Truck(Vehicle):
    def __init__(self, make, model, payload):
        """Truck 클래스 생성자: Vehicle 속성 + 적재량 초기화"""
        super().__init__(make, model)  # 부모 클래스의 생성자 호출
        self.payload = payload  # 추가 속성: 적재량

    def display_payload(self):
        """Truck 클래스의 적재량 정보 표시 메소드"""
        return f"이 트럭의 적재량은 {self.payload}ton입니다."


# Car 클래스: Vehicle 클래스를 상속받음
class Car(Vehicle):
    def __init__(self, make, model, seats):
        """Car 클래스 생성자: Vehicle 속성 + 좌석 수 초기화"""
        super().__init__(make, model)  # 부모 클래스의 생성자 호출
        self.seats = seats  # 추가 속성: 좌석 수

    def display_seats(self):
        """Car 클래스의 좌석 수 정보 표시 메소드"""
        return f"이 자동차는 {self.seats}개의 좌석을 가지고 있습니다."


# Truck 객체 생성
truck = Truck("Mercedes-Benz", "Actros MP5", 26)  # 제조사, 모델명, 적재량 초기화

# Car 객체 생성
car = Car("KIA", "SORENTO", 5)  # 제조사, 모델명, 좌석 수 초기화

# Truck 객체의 정보 출력print(truck.display_info())
print(truck.display_info())  # Vehicle 클래스의 display_info 메소드 호출
print(truck.display_payload())  # Truck 클래스의 display_payload 메소드 호출

# Car 객체의 정보 출력
print(car.display_info())  # Vehicle 클래스의 display_info 메소드 호출
print(car.display_seats())  # Car 클래스의 display_seats 메소드 호출
```
