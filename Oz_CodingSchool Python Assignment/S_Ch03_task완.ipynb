<span style="font-size:80%">[문제2]<br><br></span>
```Python의 PEP 8은 무엇인지 설명하고 PEP 8의 핵심적인 규칙을 1개 이상을 설명해주세요```

# <span style="color:red">답변</span>

PEP 8이란 "Style Guide for Python Code"로 파이썬 코드를 어떻게 구상해야할 지 알려주는 스타일 가이드이다. PEP 8은 파이썬 구문의 많은 특수성(다른 어머에서 일반적으로 적용되지 않는)을 고려하여 작성되었으며, PEP-8은 파이썬 구문을 실제로 만든 파이썬 개발자가 만들었으므로 이만큼 좋은 표준을 찾기 어렵다.

PEP 8의 주요 규칙 :

 > 1. 들여쓰기는 공백 4칸으로 사용해야 한다. 이는 코드의 구조를 명확하게 하며, 가독성을 높인다.

 > 2. 한 줄은 최대 79자까지만 작성하는 것이 좋다. 이는 코드를 여러 환경에서도 쉽게 읽을 수 있게 하며, 복잡한 표현을 간결하게 만드는 데 도움이 된다.

 > 3. 함수와 클래스, 함수 내의 큰 코드 블록 사이에는 빈 줄을 두 개 사용하여 구분한다. 이는 코드의 구조를 더 명확하게 하며, 관련 없는 코드 블록 사이의 구분을 용이하게 한다.

 > 4. 변수명과 함수명은 소문자로 작성하고, 단어 사이는 밑줄로 구분한다. 이는 이름에서 의도를 명확하게 전달하는 데 도움이 된다.

왜냐하면 이러한 규칙들은 코드의 일관성을 유지하고, 다른 개발자가 코드를 쉽게 이해할 수 있게 하기 때문이다.

<span style="font-size:80%">[문제3]<br><br></span>
```Python에서 "#"이 의마하는 것은 무엇인지 설명해주세요```

<span style="color:red">답변</span>

"#" 은 주석을 의미하며 주석이란 프로그램에 영향을 주지 않는 것으로 프로그램을 실행할 경우 무시되는 코드 영역이다.

주석은 보통 코드에서 중요한 부분에 대한 설명을 하기 위해 사용된다.

또는 테스트 중에 잠시 필요하지 않은 코드 조각을 실행하고 싶지 않을 때 해당부분을 주석으로 처리가 가능하다.

<span style="font-size:80%">[문제4]<br><br></span>
```Python의 들여쓰기 장점과 기능에 대해 설명해주세요```

<span style="color:red">답변</span>


들여쓰기의 장점과 기능 :
> 코드의 가독성 : 적절한 들여쓰기는 코드의 흐름을 명확하게 보여주므로 다른 사람들이 코드를 더 쉽게 이해할 수 있다.

> 문법적 요소 : 다른 언어에서 들여쓰기는 주로 스타일 가이드의 일부일 뿐이지만, 파이썬에서는 문법의 핵심적인 부분이다.

> 버그 예방 : 잘못된 들여쓰기는 IndentationError로 이어질 수 있어, 버그의 원인이 될수 있다.
